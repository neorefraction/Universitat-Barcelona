<h1>Problema 1: BFS (2,5p)</h1>
<p>
    Implementeu l'algorisme <b>BFS</b>. Donat un graf de la llibreria networX, la funció <b>bfs()</b> ha de retornar les caselles accessibles des del paràmetre <b>origin</b>, incloent-hi aquesta, en l'ordre d'exploració. És a dir, la llista retornada sempre inclourà, com a mínim, el node orígen.<br>
    Recordeu que és necessari <b>no canviar l'estructura de la funció</b>. No canvieu ni el nom de la funció, ni el nom dels paràmetres d'entrada ni els de sortida. <br>
</p>    
    
</div>

````python
def bfs(G, origin):
    """
    Params
    ======    
    :G: Graf on es farà la cerca. Ha de ser un objecte de la classe nx.Graph
    :origin: Índex del node orígen
    
    Returns
    =======
    :visited: El conjunt de nodes visitats durant l'exploració BFS en l'ordre en que han estat explorats
    """
    queue = [origin] # Cua on guardem els nodes a visitar
    visited = [origin] # Llista amb els nodes accessibles

    while queue: # Visitem els nodes per capes
        v = queue.pop(0)
        for u in G.neighbors(v):
            if u not in visited:
                queue.append(u)
                visited.append(u)
    
    return visited
````

````python
# Obtenim el graph associat al laberint on cada node es una cel·la i tindrem una aresta si entre dues cel·les no existeix paret
maze = Maze(7, 7, p_build=0.2, seed=8)
G = maze.maze_graph()

A = 24                    
visited = bfs(G, A)
num = len(visited)

print(f"Des de la cel·la {A} podem accedir a {num} caselles")
print(f"Ordre d'exploració: {visited}")
maze.set_values(visited, 1, color=(255,0,0))
maze.set_value(A, 1, color=(0,255,0))

maze.display_maze(plot_path=True, height=200)
````

<h1>Problema 2: Holes (2,5p)</h1>
<p>
    Implementeu una funció anomenada <b>holes()</b> que, donat un graf, un node orígen, un node destí, i un diccionari de nodes amb penalitzacions, trobi el camí més curt entre orígen i destí.<br>
    Si el camí passa per un dels nodes del diccionari, el cost de visitar el node serà la penalització en comptes de ser cost 1.<br>
    Heu de basar la vosta solució en l'algorisme <b>Dijkstra</b>.
</p>    
    
</div>

````python
def initialize_nodes(G, origin):
    """
    Inicialitza els atributs 'parent', 'distance' i 'visited' de cada node del graf, per aplicar l'algoritme Dijkstra posteriorment.
    ======
    Params
    ======
    :G: Graf on es troben els nodes que volem inicialitzar. Ha de ser un objecte de la classe nx.Graph
    :origin: Índex del node origen
    """
    G.nodes[origin].update({'distance': 0, 'parent': None, 'visited': False})

    for v in G.nodes():
        if v != origin:
            G.nodes[v].update({'distance': float('inf'), 'parent': None, 'visited': False})

def dijkstra(G, origin, destination, path, penalties):
    """
    Troba el camí minim entre un node y tots els que li son accessibles dins d'un graf no dirigit amb pesos POSITIUS
    ======
    Params
    ======
    :G: Graf en el qual es farà la cerca del camí mínim. Ha de ser un objecte de la classe nx.Graph
    :origin: Índex del node origen
    :destination: Índex del node destí
    :penalties: Diccionari amb el cost de pas dels nodes amb penalització

    Returns
    =======
    :G.nodes[u]['distance']: Cost mínim per arribar al node destí
    """

    queue = [(G.nodes[origin]['distance'], origin)]

    # Recorrem tots els nodes accessibles des del node inicial
    while queue:
        _, v = queue.pop()
        G.nodes[v]['visited'] = True
        # Calculem el cost del següent node en funció del seu pes
        for u in G.neighbors(v):
            if G.nodes[u]['visited'] == False:
                if u in penalties:
                    d = G.nodes[v]['distance'] + penalties[u]
                else:
                    d = G.nodes[v]['distance'] + 1

                # Si el nou cost és més baix que l'actual l'actualitzem i el tornem a posar a la cua
                if d < G.nodes[u]['distance']:
                    G.nodes[u].update({'distance': d, 'parent': v})
                    queue.append((G.nodes[u]['distance'], u))
                    queue.sort(reverse=True)
                
                # Si hem arribat al node final, finalitzem la cerca 
                if u == destination:
                    return G.nodes[destination]['distance']

    return float('inf')

def build_path(G, node, path):
    """
    Reconstrueix el camí mínim, d'un node A fins a un node B, després d'haver-ho cercat amb l'algoritme de Dijkstra
    ======
    Params
    ======
    :G: Graf del qual volem obtenir el path. Ha de ser un objecte de la classe nx.Graph
    :node: Índex del node a partir del qual es reconstrueix el camí
    :path: Llista amb els nodes que formen el camí
    """

    current_path = []

    # Mirem el predecessor de cada node fins a arribar al node inicial
    while node is not None:
        if node not in path:
            current_path.append(node)
        node = G.nodes[node]['parent']

    path += current_path[::-1] # Invertim la llista perquè amb Dijkstra el camí es reconstrueix des del final

def holes(G, origin, destination, holes_dct={}):
    """
    Params
    ======
    :G: Graf del qual en volem extreure el camí mínim. Ha de ser un objecte de la classe nx.Graph
    :origin: Índex del node orígen
    :destination: Índex del node destí
    :holes_dct: Un diccionari del tipus {node: penalització}. Passar per cada node del diccionari té un cost diferent a 1.
    
    Returns
    =======
    :path: Una llista de nodes del camí més curt entre els nodes 'origin' i 'destination' (ambdós inclosos).
    :cost: Un enter amb el cost de recórrer el camí, incloent-hi les penalitzacions.
    """

    # Inicialitzem tots els nodes
    path = []
    initialize_nodes(G, origin)
    cost = dijkstra(G, origin, destination, path, holes_dct) # Calculem el camí mínim des del node inicial fins el final
    build_path(G, destination, path) # Reconstruim el camí
    
    return path, cost
````

````python
# Proveu el vostre algorisme!

maze = Maze(30, 10, seed=15)

# Obtenim el graph associat al laberint on cada node es una cel·la i tindrem una aresta si entre dues cel·les no existeix paret
G = maze.maze_graph()

A = 0
B = len(G.nodes)-1
color = (255, 0, 255)

h_dcts = [
    {}, # Sense penalització
    {8:20,72:2, 227:10},
    {8:15, 72:10, 227:9},
]
for holes_dct in h_dcts:
    maze.set_values(list(holes_dct.keys()), 1, color)
    path, cost = holes(G, A, B, holes_dct)

    print(path)

    maze.set_path(path)
    print(f"Cost total: {cost}")
    print('Longitud del camí:', len(path))
    maze.display_maze(plot_path=True, height=200)

    maze.set_values(list(holes_dct.keys()), 0, color)
````

<h1>Problema 3: Checkpoint (2,5p)</h1>
<p>
    Implementeu una funció anomenada <b>checkpoint()</b> que, donat un graf, un node origen, un node destí i un node extra, trobi el camí més curt des del node origen fins el node destí passant pel node extra. Com en l'exercici anterior, tindrem un diccionari de punts on aplicarem una penalització.<br>
</p>    
    
</div>

<div class="alert alert-warning">
<p>
    Quina és la complexitat d'aquest algorisme?</b><br>
</p>    

$O((V + E) \times V) = O(V^2 + VE) = O(V^2)$

Aquest cost és a causa de la implementació feta amb un array. A l'utilitzar un array com estructura de dades auxiliar, per emmagatzemar ordenadament els nodes a visitar, és necessari controlar l'ordre dels nodes a visitar, fer això té un cost de O(V), ja que per cada node visitat s'haurà d'ordenar la llista.

Si s'usés un binary heap, en comptes d'una llista (array), el cost es veuria reduït a $O((V + E) · log(V))$ ja que en comptes d'ordenar la llista per cada node visitat, es buscaria dins del heap ($O(log(V))$) el següent. Per aconseguir-ho, es podria haver fet ús del mòdul ```heapq``` de python, que ens ofereix métodes com ```heapq.heapify()``` per crear un binary heap a partir d'una llista, ```heapq.heappush(heap, item)``` per afegir un element al heap i ```heapq.heappop(heap)``` per extreure l'element més petit.

</div>

````python
def initialize_nodes(G, origin):
    """
    Inicialitza els atributs 'parent', 'distance' i 'visited' de cada node del graf, per aplicar l'algoritme Dijkstra posteriorment.
    ======
    Params
    ======
    :G: Graf on es troben els nodes que volem inicialitzar. Ha de ser un objecte de la classe nx.Graph
    :origin: Índex del node origen
    """
    G.nodes[origin].update({'distance': 0, 'parent': None, 'visited': False})

    for v in G.nodes():
        if v != origin:
            G.nodes[v].update({'distance': float('inf'), 'parent': None, 'visited': False})

def dijkstra(G, origin, destination, path, penalties):
    """
    Troba el camí minim entre un node y tots els que li son accessibles dins d'un graf no dirigit amb pesos POSITIUS
    ======
    Params
    ======
    :G: Graf en el qual es farà la cerca del camí mínim. Ha de ser un objecte de la classe nx.Graph
    :origin: Índex del node origen
    :destination: Índex del node destí
    :penalties: Diccionari amb el cost de pas dels nodes amb penalització

    Returns
    =======
    :G.nodes[u]['distance']: Cost mínim per arribar al node destí
    """

    queue = [(G.nodes[origin]['distance'], origin)]

    # Recorrem tots els nodes accessibles des del node inicial
    while queue:
        _, v = queue.pop()
        G.nodes[v]['visited'] = True
        # Calculem el cost del següent node en funció del seu pes
        for u in G.neighbors(v):
            if u not in path and G.nodes[u]['visited'] == False:
                if u in penalties:
                    d = G.nodes[v]['distance'] + penalties[u]
                else:
                    d = G.nodes[v]['distance'] + 1

                # Si el nou cost és més baix que l'actual l'actualitzem i el tornem a posar a la cua
                if d < G.nodes[u]['distance']:
                    G.nodes[u].update({'distance': d, 'parent': v})
                    queue.append((G.nodes[u]['distance'], u))
                    queue.sort(reverse=True)
                
                # Si hem arribat al node final, finalitzem la cerca 
                if u == destination:
                    return G.nodes[destination]['distance']

    return float('inf') # Si no existeix camí tornem un cost infinit

def build_path(G, node, path):
    """
    Reconstrueix el camí mínim, d'un node A fins a un node B, després d'haver-ho cercat amb l'algoritme de Dijkstra
    ======
    Params
    ======
    :G: Graf del qual volem obtenir el path. Ha de ser un objecte de la classe nx.Graph
    :node: Índex del node a partir del qual es reconstrueix el camí
    :path: Llista amb els nodes que formen el camí
    """

    current_path = []

    # Mirem el predecessor de cada node fins a arribar al node inicial
    while node is not None:
        if node not in path:
            current_path.append(node)
        node = G.nodes[node]['parent']

    path += current_path[::-1] # Invertim la llista perquè amb Dijkstra el camí es reconstrueix des del final

def checkpoint(G, origin, destination, extra, holes_dct={}):
    """
    Params
    ======
    :G: Graf del qual en volem extreure el camí mínim. Ha de ser un objecte de la classe nx.Graph
    :origin: Índex del node orígen
    :destination: Índex del node destí
    :extra: Índex d'un node extra per on ha de passar el camí
    :holes_dct: Un diccionari del tipus {node: penalització}. Passar per cada node del diccionari té un cost diferent a 1.
    
    Returns
    =======
    :path: Una llista de nodes del camí més curt entre els nodes 'origin' i 'destination' que passa per 'extra'.
    :cost: Un enter amb el cost de recórrer el camí, incloent-hi les penalitzacions.
    """
    
    path = []
    visited = set()
    initialize_nodes(G, origin)
    cost = dijkstra(G, origin, extra, path, holes_dct) # Calculem el camí mínim per cada node des de l'origen
    if cost != float('inf'):
        build_path(G, extra, path)

    initialize_nodes(G, extra)
    cost += dijkstra(G, extra, destination, path, holes_dct) # Calculem el camí mínim per cada node des de el checkpoint
    print(cost)
    if cost != float('inf'):
        build_path(G, destination, path)
    
    return path, cost
````

````python
# Proveu el vostre algorisme!

# Creem un laberint i n'extraiem el graph
maze = Maze(10,10, seed=30)
G = maze.maze_graph()

# Definim l'inici, el fi, un node extra i la llista amb penalització
A = 30
B = len(G.nodes)-1
E = 8

h_dcts = [
    {}, # Sense penalització
    {13:3, 28:5},
    {13:10, 28:20},
]
for holes_dct in h_dcts:
    # La funció set_value ens permet 'activar' un node per a que surti dibuixat al laberint.
    maze.set_value(E, 1)                         # Checkpoint
    maze.set_values(list(holes_dct.keys()), 1, (255,0,255))  # Penalitzacions

    path, cost = checkpoint(G, A, B, E, holes_dct)
    print(f'El cost del camí més curt que passa per E és: {cost}')
    print('La distàcia del camí més curt que passa per E és:', len(path))
    print(f'El camí és: {path}')
    maze.set_path(path)
    maze.display_maze(plot_path=True, height=200)
    
    maze.set_value(E, 0)                         # Checkpoint
    maze.set_values(list(holes_dct.keys()), 0, (255,0,255))  # Penalitzacions
````

<p>

````python
from itertools import permutations

def initialize_nodes(G, origin):
    """
    Inicialitza els atributs 'parent', 'distance' i 'visited' de cada node del graf, per aplicar l'algoritme Dijkstra posteriorment.
    ======
    Params
    ======
    :G: Graf on es troben els nodes que volem inicialitzar. Ha de ser un objecte de la classe nx.Graph
    :origin: Índex del node origen
    """
    G.nodes[origin].update({'distance': 0, 'parent': None, 'visited': False})

    for v in G.nodes():
        if v != origin:
            G.nodes[v].update({'distance': float('inf'), 'parent': None, 'visited': False})

def dijkstra(G, origin, destination, path, penalties):
    """
    Troba el camí minim entre un node y tots els que li son accessibles dins d'un graf no dirigit amb pesos POSITIUS
    ======
    Params
    ======
    :G: Graf en el qual es farà la cerca del camí mínim. Ha de ser un objecte de la classe nx.Graph
    :origin: Índex del node origen
    :destination: Índex del node destí
    :penalties: Diccionari amb el cost de pas dels nodes amb penalització

    Returns
    =======
    :G.nodes[u]['distance']: Cost mínim per arribar al node destí
    """

    queue = [(G.nodes[origin]['distance'], origin)]

    # Recorrem tots els nodes accessibles des del node inicial
    while queue:
        _, v = queue.pop()
        G.nodes[v]['visited'] = True
        # Calculem el cost del següent node en funció del seu pes
        for u in G.neighbors(v):
            if u not in path and G.nodes[u]['visited'] == False:
                if u in penalties:
                    d = G.nodes[v]['distance'] + penalties[u]
                else:
                    d = G.nodes[v]['distance'] + 1

                # Si el nou cost és més baix que l'actual l'actualitzem i el tornem a posar a la cua
                if d < G.nodes[u]['distance']:
                    G.nodes[u].update({'distance': d, 'parent': v})
                    queue.append((G.nodes[u]['distance'], u))
                    queue.sort(reverse=True)
                
                # Si hem arribat al node final, finalitzem la cerca 
                if u == destination:
                    return G.nodes[destination]['distance']

    return float('inf')

def build_path(G, node, path):
    """
    Reconstrueix el camí mínim, d'un node A fins a un node B, després d'haver-ho cercat amb l'algoritme de Dijkstra
    ======
    Params
    ======
    :G: Graf del qual volem obtenir el path. Ha de ser un objecte de la classe nx.Graph
    :node: Índex del node a partir del qual es reconstrueix el camí
    :path: Llista amb els nodes que formen el camí
    """

    current_path = []

    # Mirem el predecessor de cada node fins a arribar al node inicial
    while node is not None:
        if node not in path:
            current_path.append(node)
        node = G.nodes[node]['parent']

    path += current_path[::-1] # Invertim la llista perquè amb Dijkstra el camí es reconstrueix des del final

def checkpoints_list(G, origin, destination, extras, holes_dct={}):
    """
    Params
    ======
    :G: Graf del qual en volem extreure el camí mínim. Ha de ser un objecte de la classe nx.Graph
    :origin: Índex del node orígen
    :destination: Índex del node destí
    :extras: Llista d'índexs de nodes per on ha de passar el camí.
    :holes_dct: Un diccionari del tipus {node: penalització}. Passar per cada node del diccionari té un cost diferent a 1.
    
    Returns
    =======
    :path: Una llista de nodes del camí més curt entre els nodes 'origin' i 'destination' que passa per tots els nodes 'extras'.
    :cost: Un enter amb el cost de recórrer el camí, incloent-hi les penalitzacions. Retorna float('inf') en el cas que el camí no existeixi.
    """

    path = []
    cost = float('inf')

    for checkpoints in permutations(extras): # Fem l'exploració per cada permutació posible entre els nodes extra

        current_cost = 0
        current_node = origin
        current_path = [origin]

        for v in checkpoints:
            initialize_nodes(G, current_node)
            current_cost += dijkstra(G, current_node, v, current_path, holes_dct) # Calculem el camí mínim per cada node des de el checkpoint
            build_path(G, v, current_path)
            current_node = v

        initialize_nodes(G, current_node)
        current_cost += dijkstra(G, current_node, destination, current_path, holes_dct) # Calculem el camí mínim per cada node des de el checkpoint
        build_path(G, destination, current_path)

        if current_cost < cost:
            cost = current_cost
            path = current_path

    return path, cost
````
